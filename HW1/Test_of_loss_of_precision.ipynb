```python
pa = []
pb = []
for i in range(30):
    x = 10**(-i)
    resa = (1.0 - sec(float(x)))/(tan(float(x)))**2
    resb = (1-(1-float(x))**3)/float(x)
    pa.append(resa)
    pb.append(resb)

print "Computation results using original formula a is ", pa
print "\n"
print "Computation results using original formula b is ", pb
print "\n"
output = '\n'.join('\t'.join(map(str,row)) for row in zip(pa,pb))
print output
```
